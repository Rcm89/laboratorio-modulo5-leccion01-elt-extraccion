```python
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support import expected_conditions as EC 
from selenium import webdriver 
from time import sleep
```

### Escrapeamos los datos demograficos del INE

```python
chrome_options = webdriver.ChromeOptions()


prefs = {
    "download.default_directory": r"C:\\Users\\Administrador\\Desktop\\RCM\\laboratorio-modulo5-leccion01-elt-extraccion\\data\\ine",  
    "download.prompt_for_download": False,  
    "download.directory_upgrade": True,  
    "safebrowsing.enabled": True,  
    "plugins.always_open_pdf_externally": True, 
}


url_demo = "https://www.ine.es/dyngs/INEbase/es/operacion.htm?c=Estadistica_C&cid=1254736177012&menu=resultados&idp=1254734710990"

chrome_options.add_experimental_option("prefs", prefs)

driver=webdriver.Chrome(options=chrome_options )
driver.maximize_window()
driver.get(url_demo)

# Aceptamos cookies

sleep(2)
driver.find_element("css selector", "#aceptarCookie").click()

# seleccionamo los datos que queremos por provincia

sleep(2)
driver.find_element('xpath', '/html/body/div/main/section[2]/div[1]/div[1]/div[1]/ul/li/ul/li[3]/a').click()

sleep(2)
driver.execute_script("window.scrollBy(0, 600);")

# Hacemos clic en los datos que queremos de edad etc

sleep(2)
driver.find_element("xpath", "/html/body/div/main/div[2]/ul/li[4]/ul/li[1]/a").click()

# Elegimos edades,nacionalidad,sexo
for i in range (1,4):
    driver.find_element("css selector", f"#tg{i} > div > fieldset > div.capaSelecTodosNinguno > button.opcionesvarDer > i").click()
    sleep(2)

sleep(2)
# quitamos la seleccion
driver.find_element("css selector", "#caja_periodo > div > fieldset > div.capaSelecTodosNinguno > button:nth-child(3) > i").click()

sleep(2)
# elegimos los años
for n in range(2, 5):
    WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, f'//*[@id="periodo"]/option[{n}]'))).click()
    sleep(2)


driver.execute_script("window.scrollBy(0, 600);")
sleep(2)

# Hacemos click en descargar consulta

driver.find_element("xpath", "/html/body/div/main/form/div[5]/input[3]").click()


# Hacemos click al boton de descargar

sleep(2)
driver.find_element("css selector", "#btnDescargaForm > i").click()
sleep(2)

 # Elegimos csv ;

iframe=WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "/html/body/div[6]/article/div/div/iframe")))
driver.switch_to.frame(iframe)

sleep(2)
driver.find_element("xpath", "/html/body/form/ul/li[4]/label").click()

sleep(2)
driver.switch_to.default_content()
```
